```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
dataset= pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.drop("User ID" ,axis=1)
```

```python
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
dataset.columns
```

```python
indep=dataset[[ 'Age', 'EstimatedSalary', 'Gender_Male']]
depen=dataset[['Purchased']]
```

```python
indep.shape
```

```python
depen
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(indep, depen, test_size=1/3, random_state=0)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train) # standardtion to work better block
X_test=sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier  
param_grid ={'n_neighbors':[50,10,7,5,20,58,36,96],'weights':['uniform'], 'algorithm':['auto'], 'leaf_size':[30], 'p':[2], 'metric':['minkowski'], 'n_jobs':[1]}
classifier=GridSearchCV(KNeighborsClassifier(),param_grid ,refit=True, verbose =3,n_jobs=-1,cv=5, return_train_score=False, scoring=None)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```
